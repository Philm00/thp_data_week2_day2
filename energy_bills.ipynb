```python
import pandas as pd
```

```python
df = pd.read_csv('Invoice_20201220.csv',sep="|", decimal=',')
```

### Extract the site name with a correct format

```python
df["SITE_NAME"]= df["SITE_NAME"].str.extract('.*\((.*)\).*')
```

```python
df["SITE_NAME"]
```

### ERASE invoice name

```python
df = df.drop('INVOICE_NAME', axis=1)
```

### Have the dates in a good format dd/mm/yyyy

```python
df['BEGIN'] = pd.to_datetime(df['BEGIN'], format='%d/%m/%Y')
```

```python
df['END']=pd.to_datetime(df['END'], format='%d/%m/%Y')
```

```python
df
```

```python
df['period'] = (df['END'] - df['BEGIN']).dt.days
```

```python
df
```

```python
df.dtypes
```

```python
df['PERIOD_EUR']=df['PERIOD_EUR'].str.replace(",",".")
```

```python
df['PERIOD_EUR']=df['PERIOD_EUR'].str.replace(" ","")
```

```python
df['PERIOD_EUR']=df['PERIOD_EUR'].astype("float")
```

```python
df['daily_consumption_EUR']= df['PERIOD_EUR']/df['period']
```

```python
df['KWH_PERIOD']= df['KWH_PERIOD'].str.replace(" ","")
```

```python
df['daily_consumption_KWH']= df['KWH_PERIOD'].astype("float")/df['period']
```

```python
df
```

```python
df['bills'] = df.groupby('SITE_NAME')['SITE_NAME'].transform('size')
```

```python
df
```

```python
df['CV_consumption'] = df.groupby(['SITE_NAME'])[['Expense']].transform('std') / df.groupby(['SITE_NAME'])[['Expense']].transform('mean')
```
